```python
similes = open ('output_similestat_filter.txt').readlines()
```

```python
all_e1 = []
all_e2 = []
for simile in similes:
    simile = simile.strip('\n')
    simile = simile.replace('  ',' ')
    simile, number = simile.split(' : ')
    _, e1, e2 = simile.split('as ')
    if 'the rest' not in e1 and 'the rest' not in e2:
        all_e1.append(e1)
        all_e2.append(e2)
```

```python
for e in all_e1:
    print(e)
```
